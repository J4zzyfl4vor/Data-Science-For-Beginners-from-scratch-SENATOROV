1. Python - это программное обеспечение с открытым исходным кодом. Это то
же самое, что и бесплатное ПО?
Нет. Открытый код означает доступность исходного кода для изучения и изменения, бесплатное ПО — только отсутствие платы за использование.
2. У всех ли бесплатных программ открытый исходный код? А если нет, то в чем
разница?
Нет. Бесплатное ПО может быть проприетарным (исходный код закрыт).
3. Python поддерживает динамическую типизацию. Что это такое?
Типы переменных определяются автоматически во время выполнения, а не объявляются заранее.

6. Python портативен. Что в этом контексте означает «портативность»?
Код работает на разных ОС без изменений.
7. В чем разница между «расширяемым» и «встраиваемым» языком?
    - Код работает на разных ОС без изменений.
    - Встраиваемый: можно встраивать в приложения на других языках
8. В чем смысл IDE? Чем она отличается от командной строки?
Интегрированная среда разработки — графический интерфейс с инструментами, в отличие от текстовой командной строки.
